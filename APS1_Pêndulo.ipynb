```python
#angulo 15
import math
from scipy.integrate import odeint
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt


#parâmetros do problema
teta=math.radians(15)
g=9.8 #m/s2
l = 2.17 #metros
lista_tempo = np.arange (0, 20, 0.1)

Y0 = [teta, 0]
def EqDif1(Solucoes, t):
    O=Solucoes[0]  #primeiro angulo em radiano
    W=Solucoes[1]  #zero
       
    dOdt = W
    dWdt = (-g*math.sin(O))/l
    
    return [dOdt, dWdt]

Solucao1 = odeint (EqDif1, Y0, lista_tempo)

plt.plot(lista_tempo, Solucao1[:,0]*180/math.pi,"orange")
plt.ylabel('Ângulo (graus)')
plt.xlabel('Tempo (s)')
plt.title('Ângulo por tempo IDEAL')
plt.grid(True)
plt.show()


#com aproximação sen0 = 0
#sem aproximação
#comparacao com experimental 
```
